# SolpsInterface demo
The `SolpsInterface` class from the `sims.interface` module provides a convenient interface to access and plot the results of SOLPS-ITER simulations. To create an instance of the interface, a filepath to a `balance.nc` file is given:

```python
from sims.interface import SolpsInterface
SI = SolpsInterface('balance.nc')
```

The available variables can be viewed using the `variables()` method, which returns all variable names as a list of strings:

```python
SI.variables()
```

These variable name strings are used to access and plot the SOLPS data. For example, we can plot this electron temperature using the `plot` method as follows:

```python
SI.plot('electron temperature')
```

The values of variables at a chosen set of `(R, z)` positions can be accessed using the `get()` method. First create the `R` and `z` arrays:

```python
from numpy import linspace, full
N = 20
R = full(N, fill_value=1.3)
z = linspace(-2.0, -1.8, N)
```

Now use `get()` to access the data:

```python
ne = SI.get('electron density', R, z)
```

Now plot the density values against z:

```python
import matplotlib.pyplot as plt
plt.plot(z, ne, '.-', c='blue', label='electron density')
plt.xlabel('z (m)')
plt.ylabel('electron density (m^-3)')
plt.title('lower outer-leg cross section at R = 1.3m')
plt.grid()
plt.tight_layout()
plt.show()
```
